```python
exercice4 = """
The data in JTRAIN2 come from a job training experiment conducted for low-income men during
1976–1977; see Lalonde (1986).

(i) Use the indicator variable train to determine the fraction of men receiving job training.

(ii) The variable re78 is earnings from 1978, measured in thousands of 1982 dollars. Find the
averages of re78 for the sample of men receiving job training and the sample not receiving job
training. Is the difference economically large?

(iii) The variable unem78 is an indicator of whether a man is unemployed or not in 1978. What
fraction of the men who received job training are unemployed? What about for men who did
not receive job training? Comment on the difference.

(iv) From parts (ii) and (iii), does it appear that the job training program was effective? What would
make our conclusions more convincing?
"""
print(exercice4)
```

```python
import pandas as pd
import numpy as np
import wooldridge as wl
```

```python
df = wl.data('jtrain2')
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
display(df)
```

```python
total_men = len(df)
trained_men = sum(df['train'] == 1)
ratio = ((trained_men)/(total_men))*100
```

```python
print(f'{ratio}% hommes entrainés')
```

```python
trained_men1978_earnings = df.query("train == 1")['re78'].mean()
not_trained_men1978_earnings = df.query("train == 0")['re78'].mean()

print(f'1978 trained men earnings : {trained_men1978_earnings} not trained men earnings : {not_trained_men1978_earnings}')
```

```python
ratio_earnings = ((trained_men1978_earnings - not_trained_men1978_earnings) / not_trained_men1978_earnings) * 100
print(f'En moyenne les homme qui suivent un training sont payés {ratio_earnings}% de plus que les autres')
```
